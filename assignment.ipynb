```python
#finding all patients id
upatient = list(df['patient_id'].unique())
upatient
```

```python
len(upatient)
```

```python
#converting date-time to date only
date=df['vitals_datetime']
date1 = []
for x in date:
    x = str(x)
    x= x[0:10]
    date1.append(x)
date1
```

```python
df = df.drop(["vitals_datetime"], axis=1,inplace=False)
df.head()
```

```python
df1 = pd.DataFrame(date1,columns=['date'])
df1.head()
```

```python
df = pd.concat([df1,df], axis=1)
df.head()
```

```python
df = df.drop(["Unnamed: 0"], axis=1,inplace=False)
df.head()
```

```python
#finding all unique dates
udate = list(df['date'].unique())
udate
```

```python
#Finding a High-Risk event i.e. if there is atleast one "Coded=1" for any particular patient on a day
count = 0;
hrdate=[]
pid=[]
for y in upatient:
    l=[]
    for i in range(len(df)):
        if df['patient_id'][i]==y:
            l.append([df['date'][i],df["Coded"][i]])
    a=[]
    b=[]
    for x in l:
        if x[1]==1:
            a.append(x[0])
    b.append(y)
    count=count+1
    print(count)
    if list(set(a))!=[]:
        print("patient_id =",list(set(b)),"is on high risk on",list(set(a)))
    else:
        print("patient_id =",list(set(b)),"is never on high risk")
    hrdate.append(list(set(a)))
    pid.append((set(b)))
```

```python
h=pd.DataFrame()
for i in range(len(hrdate)):
    list=hrdate[i]
    if list==[]:
        list=['Never on high risk']
        h=pd.concat([pd.DataFrame([[list]],columns=['Date']),h])
    else:
        h=pd.concat([pd.DataFrame([[list]],columns=['Date']),h])
    
h.head()
```

```python
df2 = pd.DataFrame(pid,columns=['patient_id'])
df2
```

```python
h.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

h = pd.concat([df2,h],axis=1)
h.head()
```

```python
h['High Risk Date'] = [','.join(map(str, l)) for l in h['Date']]
h.head()
```

```python
h = h.drop(["Date"], axis=1,inplace=False)
h.head()
```

```python
#Final file of High-Risk event i.e. if there is atleast one "Coded=1" for any particular patient and on what day
h.to_csv(r'High_Risk.csv', index = False)
```

```python
#Import models from scikit learn module:
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics

#Generic function for making a classification model and accessing performance:
def classification_model(model, data, predictors, outcome):
      model.fit(data[predictors],data[outcome])
      predictions = model.predict(data[predictors])
      print(predictions)
      accuracy = metrics.accuracy_score(predictions,data[outcome])
      print(accuracy*100)
```

```python
#logistic Regression
outcome_var=['Coded']
predictor_var = ['heart_rate', 'respiration_over_impedence','spirometry_oxygen_saturation', 'pulse', 'blood_pressure_systolic','blood_pressure_diastolic', 'blood_pressure_average','machine_id']
model =LogisticRegression()
classification_model(model, df,predictor_var,outcome_var)
```

```python
#Gaussian NAive bayes algorithm
from sklearn.naive_bayes import GaussianNB
outcome_var=['Coded']
predictor_var = ['heart_rate', 'respiration_over_impedence','spirometry_oxygen_saturation', 'pulse', 'blood_pressure_systolic','blood_pressure_diastolic', 'blood_pressure_average','machine_id']
model = GaussianNB() 
classification_model(model, df,predictor_var,outcome_var)
```

```python
#applying knn algorithm
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
classification_model(model, df,predictor_var,outcome_var)
```

```python
#Gradient boost algorithm
from sklearn.ensemble import GradientBoostingClassifier
model= GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
classification_model(model, df,predictor_var,outcome_var)
```

```python
#Decision tree algorithm
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion='entropy')
classification_model(model, df,predictor_var,outcome_var)
```

```python
#Random forest classifier
model = RandomForestClassifier()
classification_model(model, df,predictor_var,outcome_var)
```

```python
#Loading and processing training data
training_data=pd.read_csv('training_frame.csv')
training_data = training_data.drop(["Unnamed: 0","vitals_datetime","patient_id","machine_id"], axis=1,inplace=False)
training_data.head()
```

```python
#allocating training data
X_train = training_data.iloc[:, 0:7]
y_train = training_data.iloc[:, 7:8]
```

```python
#Loading and processing test data
newdata=pd.read_csv('test_frame.csv')
newdata = newdata.drop(["Unnamed: 0","vitals_datetime","patient_id","machine_id"], axis=1,inplace=False)
newdata.head()
```

```python
#allocating test data
X_test = newdata.iloc[:, 0:7]
y_test = newdata.iloc[:, 7:8]
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Neural Network Model Training
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
mlp.fit(X_train, y_train.values.ravel())
```

```python
#predictions of Neural Network data
predictions = mlp.predict(X_test)
```

```python
#Evaluating the model
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
